```python
from fastai.text.all import *
```

imported the fastai text module

```python
path = untar_data(URLs.IMDB)
```

created a path for data from IMDB dataset 

```python
dls = TextDataLoaders.from_folder(path,valid='test')
```

loaded the data from the dataset to a "dls" object of text specific dataloaders

```python
dls.show_batch()
```

getting a view of the batch

```python
!pip install -U fastai fastcore
learn = text_classifier_learner(dls,AWD_LSTM,drop_mult=0.5,metrics = accuracy)
with learn.no_bar():
     learn.fine_tune(4,1e-2)
```

installed the fastcore module
created a learner object "learn" to train the text model
fine tunes the models upto 4 epochs 


```python
learn.predict("the only good thing about this movie was the 'Interval' because it finally gave me a break from the torture")
```
